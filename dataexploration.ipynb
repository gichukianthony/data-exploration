```python
data = [50,50,47,97,49,3,53,42,26,74,82,62,37,
        15,70,27,36,35,48,52,63,64]

# Install numpy if not already installed
# %pip install numpy
                                           #USING NUMPY
import numpy as np
grades = np.array(data)
print (grades)
#difference between a list an an numpy array
print(type(data), 'x 2:', data * 2)
print ('---')
print (type(grades), 'x 2:', grades * 2)

#shape of our array ie no of elements
grades.shape

#accesing elements at diff indexes
grades[0]

#reversing the array
grades[::-1]

#sorting the array
grades.sort()

#mean of the array

grades.mean()

study_hours= [10.0,11.5,9.0,16.0,9.25,1.0,11.5,9.0,8.5,14.5,15.5,
               13.75,9.0,8.0,15.5,8.0,9.0,6.0,10.0,12.0,12.5,12.0]

#creating a 2d array
student_data = np.array([study_hours, grades])
print (student_data)

student_data.shape

student_data[0] [0]

                                     #USING PANDAS
import pandas as pd
# %pip install 
df_students = pd.DataFrame({'Name': ['Dan','Joann','Pedro', 'Rosie', 'Ethan', 'Vicky', 'Frederic', 'Jimmie', 
                                     'Rhonda', 'Giovanni', 'Francesca', 'Rajab', 'Naiyana', 'Kian', 'Jenny',
                                     'Jakeem','Helena','Ismat','Anila','Skye','Daniel','Aisha'],
                                'studyHours':student_data[0],
                                 'Grades':student_data[1]})
df_students

df_students.describe()

df_students.loc[5]

df_students.sort_values(by='Grades', ascending=False)
df_students.loc[0:5]#position
df_students.iloc[0:5]#index

df_students.iloc[0,[1,2]]

df_students.loc[0,'Grades']

df_students.loc[df_students['Name']=='Aisha']

df_students.query('Name=="Aisha"')

                        #LOADING DATA FROM A FILE
df_students= pd.read_csv('./data/grades.csv',delimiter=',',header='infer')
df_students.head()

df_students.isnull()
df_students.isnull().sum()
df_students[df_students.isnull().any(axis=1)]

df_students.dropna(inplace=True)
df_students=df_students.dropna(axis=0, how='any')
df_students

      #EXPLORING DATA IN DATA FRAMES
#getting the mean study hours using the column name as an index
mean_study= df_students['StudyHours'].mean()

#mean grade as  a property

mean_grade = df_students.Grade.mean()

print ('average weekly study hours :{:.2f} \naverage grade :{:.2f}'.format(mean_study,mean_grade))

passes = pd.Series(df_students['Grade']>=60)
df_students= pd.concat([df_students,passes.rename("Pass")], axis= 1)
df_students

print(df_students.groupby(df_students.Pass).Name.count())

#VISUALIZING DATA USING MATPLOTLIB
%matplotlib inline

from matplotlib import pyplot as plt
#bar plot
plt.bar(x=df_students.Name, height=df_students.Grade)
plt.show()

plt.bar(x=df_students.Name, height=df_students.Grade, color='orange')
plt.title('students grades')
plt.xlabel('Student')
plt.ylabel('Grade')
plt.grid(color='#95a5a6', linestyle='--',linewidth= 2, axis='y', alpha=0.7)
plt.xticks(rotation=90)

plt.show()
```
